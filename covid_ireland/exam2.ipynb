```python
import pandas as pd
import bokeh.plotting as bp
import bokeh.models as bm
import json
```

```python
df = pd.read_csv('data/munged/df.csv')
df.head()
```

```python
with open('data/counties.json') as f: counties = json.load(f)
```

```python
df['color'] = df['CountyName'].apply(lambda x: counties[x]['color'])
```

```python
bp.output_notebook()
```

```python
tools = bm.HoverTool(tooltips=[("County", "@CountyName"), ("Date", "@TimeStamp")])
```

```python
p = bp.figure(width=800, height=800, title="Cases v Density")
source = bm.ColumnDataSource(df)
p.circle('CaseDensity', 'ConfirmedCovidCases', fill_color='color', line_color=None, source=source)
p.background_fill_color="bisque"
p.xaxis.axis_label = 'Case Density'
p.yaxis.axis_label = 'Confirmed Cases'
p.add_tools(tools)
bp.show(p)
```
